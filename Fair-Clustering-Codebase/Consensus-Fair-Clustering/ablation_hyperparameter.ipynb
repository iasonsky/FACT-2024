```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize a dictionary to store the results
results = { 'balance': [], 'entropy': [], 'nmi': [], 'acc': [] }

for hyperparam, values in hyperparameters.items():
    for value in values:
        command = f'python train.py --no-cuda --{hyperparam} {value}'
        print(f'Running: {command}')
        
        output = !{command}
        
        output_json = json.loads(output[0])
        
        for key in results.keys():
            results[key].append(output_json[key])

df = pd.DataFrame(results)

for result in results.keys():
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[[result]], annot=True, cmap='viridis')
    plt.title(f'Heatmap for {result}')
    plt.show()
```
