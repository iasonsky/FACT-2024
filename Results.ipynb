```python
import numpy as np
import pickle
```

```python
def print_results(result_dict, result_name, name, cl_algo):
    '''Prints and saves the results in a pickle file.'''
    print("")
    print(f'{result_name} Results for {name} and {cl_algo}')
    print("=========================================")
    for metric in result_dict.keys():
        if "N/A" in result_dict[metric]:
            print(f'{metric}: {result_dict[metric]}')
            continue
        mean_val = np.mean(result_dict[metric])
        std_val = np.std(result_dict[metric])
        print(f'{metric}: Mean = {mean_val}, Std = {std_val}')
```

```python
path_name = "Fair-Clustering-Codebase/extra_metrics_results/Yale/KFC/"
post_attack_res = pickle.load(open(path_name + "post_attack_res.pkl", "rb"))
pre_attack_res = pickle.load(open(path_name + "pre_attack_res.pkl", "rb"))
random_attack_res = pickle.load(open(path_name + "random_attack_res.pkl", "rb"))

path_name_old = "Fair-Clustering-Codebase/extra_metrics_results/Yale/KFC/old/"
post_attack_res_old = pickle.load(open(path_name_old + "post_attack_res.pkl", "rb"))
pre_attack_res_old = pickle.load(open(path_name_old + "pre_attack_res.pkl", "rb"))
random_attack_res_old = pickle.load(open(path_name_old + "random_attack_res.pkl", "rb"))
```

```python
print("Pre-Attack Results")
print("==================")
print_results(pre_attack_res, "Pre-Attack", "Yale", "KFC")
print_results(pre_attack_res_old, "Old Pre-Attack", "Yale", "KFC")

print("Post-Attack Results")
print("===================")
print_results(post_attack_res, "Post-Attack", "Yale", "KFC")
print_results(post_attack_res_old, "Old Post-Attack", "Yale", "KFC")

print("Random-Attack Results")
print("=====================")
print_results(random_attack_res, "Random-Attack", "Yale", "KFC")
print_results(random_attack_res_old, "Old Random-Attack", "Yale", "KFC")

```
